```python
# importing libraries
import numpy as np
import operator
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
# for calculating mean_squared error
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

```python
# 10 points from the noisy sin(2*pi*x)
import numpy as np

x = np.random.random_sample(5,)
y = np.sin(2*np.pi*x)+0.2*np.random.normal(0,0.05,5) #Add noise of mean 0 and standard deviation 1 
x1 = np.random.random_sample(10,)
y1 = np.sin(2*np.pi*x1)+0.2*np.random.normal(0,0.05,10) #Add noise of mean 0 and standard deviation 1 
x2 = np.random.random_sample(50,)
y2 = np.sin(2*np.pi*x2)+0.2*np.random.normal(0,0.05,50) #Add noise of mean 0 and standard deviation 1 
x3 = np.random.random_sample(100,)
y3 = np.sin(2*np.pi*x3)+0.2*np.random.normal(0,0.05,100) #Add noise of mean 0 and standard deviation 1 
x4 = np.random.random_sample(500,)
y4 = np.sin(2*np.pi*x4)+0.2*np.random.normal(0,0.05,500) #Add noise of mean 0 and standard deviation 1 
x5 = np.random.random_sample(1000,)
y5 = np.sin(2*np.pi*x5)+0.2*np.random.normal(0,0.05,1000) #Add noise of mean 0 and standard deviation 1 
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
X1_train, X1_test, y1_train, y1_test = train_test_split(x1, y1, test_size = 0.2)
X2_train, X2_test, y2_train, y2_test = train_test_split(x2, y2, test_size = 0.2)
X3_train, X3_test, y3_train, y3_test = train_test_split(x3, y3, test_size = 0.2)
X4_train, X4_test, y4_train, y4_test = train_test_split(x4, y4, test_size = 0.2)
X5_train, X5_test, y5_train, y5_test = train_test_split(x5, y5, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# importing libraries for polynomial transform
from sklearn.preprocessing import PolynomialFeatures
# for creating pipeline
from sklearn.pipeline import Pipeline
# creating pipeline and fitting it on data
Input=[('polynomial',PolynomialFeatures(degree=10)),('modal',LinearRegression())]
pipe=Pipeline(Input)
pipe.fit(X_train.reshape(-1,1),y_train.reshape(-1,1))
pipe.fit(X1_train.reshape(-1,1),y1_train.reshape(-1,1))
pipe.fit(X2_train.reshape(-1,1),y2_train.reshape(-1,1))
pipe.fit(X3_train.reshape(-1,1),y3_train.reshape(-1,1))
pipe.fit(X4_train.reshape(-1,1),y4_train.reshape(-1,1))
pipe.fit(X5_train.reshape(-1,1),y5_train.reshape(-1,1))
```

```python
poly_pred=pipe.predict(X_train.reshape(-1,1))
#sorting predicted values with respect to predictor
sorted_zip = sorted(zip(X_train,poly_pred))
x_poly, poly_pred = zip(*sorted_zip)
test_prediction=pipe.predict(X_test.reshape(-1,1))
MSE=mean_squared_error(y_test,test_prediction)

#plotting predictions
plt.figure(figsize=(10,6))
plt.scatter(X_train,y_train,label='Training Points',s=15)
plt.plot(x_poly,poly_pred,color='r',label='Polynomial Regression')
plt.title("MSE = {}\n Sample = {}".format(MSE,'5'))
plt.legend()
plt.show()
```
